```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
data=pd.read_csv("C:\\Users\\cemre\\Desktop\\Dry_Bean_Dataset.csv")
```

```python
features = data.drop('Class', axis=1)
labels = data['Class']
```

```python
# Split the dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)
```

```python
# Create a KNN classifier with k=10
classifier = KNeighborsClassifier(n_neighbors=10)

# Train the classifier
classifier.fit(X_train, y_train)
```

```python
# Generate predictions
y_pred = classifier.predict(X_test)
```

```python
# Calculate  accuracy, precision, recall, and F-score 
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f_score = f1_score(y_test, y_pred, average='macro')

print("KNN Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F-score:", f_score)
```

```python
# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# Create a bar plot for the confusion matrix
plt.figure(figsize=(10, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix for KNN")
plt.colorbar()
tick_marks = np.arange(len(np.unique(labels)))
plt.xticks(tick_marks, np.unique(labels), rotation=45)
plt.yticks(tick_marks, np.unique(labels))
plt.xlabel('Predicted Class')
plt.ylabel('True Class')

# Add value labels to the plot
thresh = cm.max() / 2
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], 'd'), ha='center', va='center',
             color='white' if cm[i, j] > thresh else 'black')

plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data=pd.read_csv("C:\\Users\\cemre\\Desktop\\Dry_Bean_Dataset.csv")

# Scatter plot of Area and Perimeter
plt.scatter(data['Area'], data['Perimeter'])
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.title('Scatter Plot of Area vs. Perimeter')
plt.show()
```

```python

# Histogram of Aspect Ratio
plt.hist(data['AspectRatio'], bins=20)
plt.xlabel('Aspect Ratio')
plt.ylabel('Frequency')
plt.title('Histogram of Aspect Ratio')
plt.show()
```

```python

# Bar chart of Bean Type counts
type_counts = data['Class'].value_counts()
plt.bar(type_counts.index, type_counts.values)
plt.xlabel('Bean Type')
plt.ylabel('Count')
plt.title('Bean Type Counts')
plt.show()
```

```python

# Box plot of Compactness by Bean Type
sns.boxplot(x='Class', y='Compactness', data=data)
plt.xlabel('Bean Type')
plt.ylabel('Compactness')
plt.title('Box Plot of Compactness by Bean Type')
plt.show()
```

```python

# Pair plot of selected features
selected_features = ['Area', 'Perimeter', 'Compactness']
sns.pairplot(data[selected_features])
plt.title('Pair Plot of Selected Features')
plt.show()
```
